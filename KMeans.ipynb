```python
x=country.iloc[:,[3,4]].values
```

```python
print(x)
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters=6,init='k-means++',random_state=0)
y=kmeans.fit_predict(x)
```

```python
print(y)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0],x[y==0,1],s=50,c='blue',label='Cluster1')
plt.scatter(x[y==1,0],x[y==1,1],s=50,c='green',label='Cluster2')
plt.scatter(x[y==2,0],x[y==2,1],s=50,c='pink',label='Cluster3')
plt.scatter(x[y==3,0],x[y==3,1],s=50,c='black',label='Cluster4')
plt.scatter(x[y==4,0],x[y==4,1],s=50,c='gray',label='Cluster5')
plt.scatter(x[y==5,0],x[y==5,1],s=50,c='yellow',label='Cluster6')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='red',label='Centroids')
plt.title("Country Group")
plt.xlabel('GDP')
plt.show()
```
